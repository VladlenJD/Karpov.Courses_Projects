```python


Задача:

В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. 
В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. 
Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно 
оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной 
попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик 
оценки результата? Если да, то как?

```

```python


Механика проведения события №1.


1) Involvement_metric(вовлечённость) = EAU (event active users) / MAU. Посмотрим, какая часть пользователей, активных 
   хотя бы раз в месяц, заинтересовались нашим событием(зарегистрировались или начали проходить).

2) Winner_metric = количество пользователей, успешно прошедших событие / EAU. Посмотрим, какая часть зарегистрировавшихся
   на событие пользователей дошла до конца. Если слишком мало, пересмотрим количество уровней для прохождения(даже занимаясь
   садоводством, можно быть нетерпеливым:)), пересмотрим сложность уровней. Если же слишком много, то возможно стоит усложнить 
   уровни или увеличить их количество, чтобы оставался спортивный интерес, и пользователи не заскучали(при этом улучшим приз: 
   сделать саженцы не 40 уровня, а 60, чтобы вызвать больший интерес к прохождению до конца, а если анонсировать призы заранее,
   то наша первая метрика Involvement_metric тоже увеличится).

3) ATPP (average time per passage) = event.time.mean(). Посмотрим на среднее время прохождения события, не слишком ли быстро
   пользователи проходят все уровни, не стоит ли нам усложнить?
    
4) TPE(time per event) = ATTP / ATIA(average time in app). Посмотрим, какую часть времени своего нахождения в приложении, 
   пользователи(EAU) в среднем тратили на прохождения события. Если метрика стремится к 1(или > 1), то пользователи заходили 
   ради события. Это хорошо характеризует само событие.

5) Ignore_rate = EAU / users_in_app. Метрика показывает, какая часть пользователей зашла в приложение во время события, но
   не стала даже регистрироваться в нём. Может быть им нравится поливать свой сад, но наше событие 
   по каким-то причинам не понравилось. Нужно разбираться, почему.
    
6) ER (event revenue) = Общий доход, полученный с пользователей, зарегистрировавшихся на событие. Можем изначально подобрать 
   сложность уровней так, чтобы какая-то часть MAU не могла пройти, к примеру, последний уровень без какого-нибудь хитрого 
   артефакта, который можно прикупить в магазине за символические деньги(чтобы повысить конверсию в покупку).

```
